## Visualise the data

#### Compare the number of disaster and regular tweets

```python
positive_no = len(train[train['target'] == 1])
negative_no = len(train[train['target'] == 0])

plt.bar(['Disaster', 'Regular'], [positive_no, negative_no], color = ['blue', 'green'], alpha = 0.5, edgecolor = 'black')
plt.title('Number of tweets per class')
plt.show()
```

#### Compare the number of characters per tweet

```python
positive_character_no = [len(text) for text in train[train["target"] == 1]["text"]]
negative_character_no = [len(text) for text in train[train["target"] == 0]["text"]]

num_bins = 30
_, bins, _ = plt.hist(positive_character_no, num_bins, alpha = 0.5, label = "disaster tweets", facecolor = 'blue', edgecolor = 'black')
plt.hist(negative_character_no, bins, alpha = 0.5, label = "regular tweets", facecolor = 'green', edgecolor = 'black')
plt.xlabel("Character no")
plt.ylabel("Tweet no")
plt.title("Number of characters per tweet - train set")
plt.legend(loc = 'upper left')
plt.show()
```

#### Compare the number of words per tweet

```python
positive_word_no = train[train['target'] == 1]["text"].apply(lambda tweet : len(tweet.split(' ')))
negative_word_no = train[train['target'] == 0]['text'].apply(lambda tweet : len(tweet.split(' ')))

_, bins, _ = plt.hist(positive_word_no, num_bins, alpha = 0.5, label = 'disaster tweets', facecolor = 'blue', edgecolor = 'black')
plt.hist(negative_word_no, bins, alpha = 0.5, label = 'regular tweets', facecolor = 'green', edgecolor = 'black')
plt.xlabel("Word no")
plt.ylabel("Tweet no")
plt.title("Number of words per tweet - train set")
plt.legend()
plt.show()
```

#### Compate the average word length

```python
positive_avg_word = positive_character_no / positive_word_no
negative_avg_word = negative_character_no / negative_word_no

_, bins, _ = plt.hist(positive_avg_word[positive_avg_word < 50], num_bins, alpha = 0.5, facecolor = 'blue', edgecolor = 'black', label = 'disaster tweets') # Remove outliers
plt.hist(negative_avg_word[negative_avg_word < 50], bins, alpha = 0.5, facecolor = 'green', edgecolor = 'black', label = 'regular tweets') # Remove outliers
plt.xlabel("Average word length")
plt.ylabel("Tweet no")
plt.title("Average word length per tweet - train set")
plt.legend()
plt.show()
```

#### Compare the average number of tweets with uppercase words between classes

```python
positive_uppercase = train[train['target'] == 1]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.isupper()]))
negative_uppercase = train[train['target'] == 0]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.isupper()]))
positive_uppercase_avg = len(positive_uppercase[positive_uppercase > 0]) / len(positive_uppercase)
negative_uppercase_avg = len(negative_uppercase[negative_uppercase > 0]) / len(negative_uppercase)

plt.bar(['Disaster', 'Regular'], [positive_uppercase_avg, negative_uppercase_avg], color = ['blue', 'green'], alpha = 0.5, edgecolor = 'black')
plt.ylabel('Avg number of uppercase tweets')
plt.show()
```

#### Compare the average number of urls between classes

```python
# Preprocessing step: replace links in tweets with 'url' using regex
url_regex = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
train_without_urls = train['text'].apply(lambda tweet : re.sub(url_regex, 'url', tweet))

positive_urls = train_without_urls[train['target'] == 1].apply(lambda tweet : len([word for word in tweet.split() if word == "url"]))
positive_url_avg = positive_urls.sum() / len(positive_urls)

negative_urls = train_without_urls[train['target'] == 0].apply(lambda tweet : len([word for word in tweet.split() if word == "url"]))
negative_url_avg = negative_urls.sum() / len(negative_urls)

plt.bar(['Disaster', 'Regular'], [positive_url_avg, negative_url_avg], color = ['blue', 'green'], alpha = 0.5, edgecolor = 'black')
plt.ylabel('Avg number of urls')
plt.show()
```

## Preprocess the Data

Basic preprocessing

```python
# Preprocessing steps chosen to optimise the predictions
def preprocessing(dataset):
    # Convert tweets to lowercase
    dataset['text'] = dataset['text'].apply(lambda tweet : " ".join(word.lower() for word in tweet.split())) 
    
    # Change urls in tweets to 'url' word
    #url_regex = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    #dataset['text'] = dataset['text'].apply(lambda tweet : re.sub(url_regex, 'url', tweet))
    
    # Remove punctuation
    dataset['text'] = dataset['text'].str.replace('[^\w\s]', '') 

    # Spelling correction changes names of places that could contribute to accurate prediction
    #dataset['text'][:10].apply(lambda tweet : str(TextBlob(tweet).correct()))    

    # Lemmatization of the words
    dataset['text'] = dataset['text'].apply(lambda tweet : " ".join(Word(word).lemmatize() for word in tweet.split())) 

    # Remove stop words from tweets
    #stop = stopwords.words('english') # Get english stopwords list
    #dataset['text'] = dataset['text'].apply(lambda tweet : " ".join(word for word in tweet.split() if word not in stop)) 

    # Rarest / Most common words
    # freq = pd.Series(' '.join(train['text']).split()).value_counts()[-50:]
    
    return dataset
    
```

Sentiment analysis

```python
def sentiment_analysis(dataset):
    # Append a vector of sentiment polarity (-1 - 1 range)
    dataset['sentiment'] = dataset['text'].apply(lambda tweet : TextBlob(tweet).sentiment[0])
    
    return dataset
```

Apply preprocessing to the train dataset

```python
preprocessing(train) # Passed by reference, use train[:] to pass by value

#sentiment_analysis(train)
```

## Create the model

Extract the token count

```python
count_vectoriser = feature_extraction.text.CountVectorizer(ngram_range=(1, 3))
tokens = count_vectoriser.fit_transform(train["text"])
```

Fit the model

```python
model = MultinomialNB(alpha = 5)
#model = SVC(C = 5, kernel = 'poly')

# Naive Bayes
training_data = tokens

# SVM
#training_data = hstack((tokens, train[['sentiment']].values), format='csr') # Combine tokens with other features
#columns = count_vectoriser.get_feature_names()+train[['sentiment']].columns.tolist() # Columns of newly created matrix

# Cross validation
test_score = model_selection.cross_val_score(model, training_data, train["target"], cv = 5, scoring = "accuracy")
print('Cross validation acc: ', test_score.mean())

# Fit the model
model.fit(training_data, train["target"])
print("Mean training acc: ", model.score(training_data, train["target"]))
```

#### Submission

```python
def submission(submission_file_path, model, test_vectors):
    sample_submission = pd.read_csv(submission_file_path)
    sample_submission["target"] = model.predict(test_vectors)
    sample_submission.to_csv("submission.csv", index=False)
```

```python
submission_file_path = "Data/sample_submission.csv"

# Apply preprocessing to test set
preprocessing(test)
#sentiment_analysis(test)

test_vectors = count_vectoriser.transform(test['text'])

# Naive Bayes
test_data = test_vectors

# SVM
#sentiment_analysis(test)
#test_data = hstack((test_vectors, test[['sentiment']].values), format='csr')
```

```python
submission(submission_file_path, model, test_data)
```
