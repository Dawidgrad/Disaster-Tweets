## Import everything

Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn import feature_extraction, linear_model, model_selection
from nltk.corpus import stopwords
from textblob import TextBlob
```

Import the dataset

```python
train = pd.read_csv("Data/train.csv")
test = pd.read_csv("Data/test.csv")
```

## Visualise the data

#### Compare the number of disaster and regular tweets

```python
positive_no = len(train[train['target'] == 1])
negative_no = len(train[train['target'] == 0])

plt.bar(['Disaster', 'Regular'], [positive_no, negative_no], color = ['blue', 'green'], alpha = 0.5, edgecolor = 'black')
plt.title('Number of tweets per class')
plt.show()
```

#### Compare the number of characters per tweet

```python
positive_character_no = [len(text) for text in train[train["target"] == 1]["text"]]
negative_character_no = [len(text) for text in train[train["target"] == 0]["text"]]

num_bins = 30
_, bins, _ = plt.hist(positive_character_no, num_bins, alpha = 0.5, label = "disaster tweets", facecolor = 'blue', edgecolor = 'black')
plt.hist(negative_character_no, bins, alpha = 0.5, label = "regular tweets", facecolor = 'green', edgecolor = 'black')
plt.xlabel("Character no")
plt.ylabel("Tweet no")
plt.title("Number of characters per tweet - train set")
plt.legend(loc = 'upper left')
plt.show()
```

#### Compare the number of words per tweet

```python
positive_word_no = train[train['target'] == 1]["text"].apply(lambda tweet : len(tweet.split(' ')))
negative_word_no = train[train['target'] == 0]['text'].apply(lambda tweet : len(tweet.split(' ')))

_, bins, _ = plt.hist(positive_word_no, num_bins, alpha = 0.5, label = 'disaster tweets', facecolor = 'blue', edgecolor = 'black')
plt.hist(negative_word_no, bins, alpha = 0.5, label = 'regular tweets', facecolor = 'green', edgecolor = 'black')
plt.xlabel("Word no")
plt.ylabel("Tweet no")
plt.title("Number of words per tweet - train set")
plt.legend()
plt.show()
```

#### Compate the average word length

```python
positive_avg_word = positive_character_no / positive_word_no
negative_avg_word = negative_character_no / negative_word_no

_, bins, _ = plt.hist(positive_avg_word[positive_avg_word < 50], num_bins, alpha = 0.5, facecolor = 'blue', edgecolor = 'black', label = 'disaster tweets') # Remove outliers
plt.hist(negative_avg_word[negative_avg_word < 50], bins, alpha = 0.5, facecolor = 'green', edgecolor = 'black', label = 'regular tweets') # Remove outliers
plt.xlabel("Average word length")
plt.ylabel("Tweet no")
plt.title("Average word length per tweet - train set")
plt.legend()
plt.show()
```

#### Compare number of numericals between the classes

```python
positive_numericals = train[train['target'] == 1]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.isdigit()]))
negative_numericals = train[train['target'] == 0]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.isdigit()]))

bins_num = 5
_, bins, _ = plt.hist(positive_numericals[positive_numericals > 0], bins_num, alpha = 0.5, facecolor = 'blue', edgecolor = 'black', label = 'disaster tweets')
plt.hist(negative_numericals[negative_numericals > 0], bins, alpha = 0.5, facecolor = 'green', edgecolor = 'black', label = 'regular tweets')
plt.xlabel("Number of numericals")
plt.ylabel("Number of tweets")
plt.title("Number of numericals per tweet compared between classes")
plt.legend()
plt.show()
```

#### Compare the number of hashtags between classes

```python
positive_hashtag_no = train[train['target'] == 1]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.startswith('#')]))
negative_hashtag_no = train[train['target'] == 0]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.startswith('#')]))

bins_num = 10
_, bins, _ = plt.hist(positive_hashtag_no, bins_num, alpha = 0.5, facecolor = 'blue', edgecolor = 'black', label = 'disaster tweets')
plt.hist(negative_hashtag_no, bins, alpha = 0.5, facecolor = 'green', edgecolor = 'black', label = 'regular tweets')
plt.xlabel('Hashtags no')
plt.ylabel('Tweets no')
plt.title('Number of hashtags between classes')
plt.legend()
plt.show()
```

#### Compare the number of tweets with uppercase words between classes

```python
positive_uppercase = train[train['target'] == 1]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.isupper()]))
negative_uppercase = train[train['target'] == 0]['text'].apply(lambda tweet : len([tweet for unigram in tweet.split() if unigram.isupper()]))
positive_uppercase_no = len(positive_uppercase[positive_uppercase > 0])
negative_uppercase_no = len(negative_uppercase[negative_uppercase > 0])

plt.bar(['Disaster', 'Regular'], [positive_uppercase_no, negative_uppercase_no], color = ['blue', 'green'], alpha = 0.5, edgecolor = 'black')
plt.show()
```

## Preprocess the Data

```python
train['text'] = train['text'].apply(lambda tweet : " ".join(word.lower() for word in tweet.split())) # Convert tweets to lowercase
train['text'] = train['text'].str.replace('[^\w\s]', '') # Remove punctuation (should # be removed?)

stop = stopwords.words('english') # Get english stopwords list

train['text'] = train['text'].apply(lambda tweet : " ".join(word for word in tweet.split() if word not in stop)) # Remove stop words from tweets
```

## Create the model

Extract the token count

```python
count_vectoriser = feature_extraction.text.CountVectorizer(ngram_range=(2,2))
tokens = count_vectoriser.fit_transform(train["text"])
```

```python
test.isnull().sum()
```

Fit the model

```python
model = linear_model.RidgeClassifier()
scores = model_selection.cross_val_score(model, tokens, train["target"], cv = 3, scoring = "f1")
model.fit(tokens, train["target"])
```
