## Prepare everything

Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import feature_extraction, linear_model, model_selection
```

Import the dataset

```python
train = pd.read_csv("Data/train.csv")
test = pd.read_csv("Data/test.csv")
```

## Visualise the data

Compare the number of disaster and regular tweets

```python
positive_no = len(train[train['target'] == 1])
negative_no = len(train[train['target'] == 0])
plt.bar(['Disaster', 'Regular'], [positive_no, negative_no], color = ['blue', 'green'], alpha = 0.5, edgecolor = 'black')
plt.title('Number of tweets per class')
plt.show()
```

Find out the number of characters per tweet

```python
positive_character_no = [len(text) for text in train[train["target"] == 1]["text"]]
negative_character_no = [len(text) for text in train[train["target"] == 0]["text"]]
num_bins = 30
plt.hist(positive_character_no, num_bins, alpha = 0.5, label = "disaster tweets", facecolor = 'blue', edgecolor = 'black')
plt.hist(negative_character_no, num_bins, alpha = 0.5, label = "regular tweets", facecolor = 'green', edgecolor = 'black')
plt.xlabel("Character no")
plt.ylabel("Tweet no")
plt.title("Number of characters per tweet - train set")
plt.legend(loc = 'upper left')
plt.show()
```

Find out the number of words per tweet

```python
positive_word_no = train[train['target'] == 1]["text"].apply(lambda tweet : len(tweet.split(' ')))
negative_word_no = train[train['target'] == 0]['text'].apply(lambda tweet : len(tweet.split(' ')))
plt.hist(positive_word_no, num_bins, alpha = 0.5, label = 'disaster tweets', facecolor = 'blue', edgecolor = 'black')
plt.hist(negative_word_no, num_bins, alpha = 0.5, label = 'regular tweets', facecolor = 'green', edgecolor = 'black')
plt.xlabel("Word no")
plt.ylabel("Tweet no")
plt.title("Number of words per tweet - train set")
plt.legend()
plt.show()
```

Average word length

```python
positive_avg_word = positive_character_no / positive_word_no
negative_avg_word = negative_character_no / negative_word_no
plt.hist(positive_avg_word[positive_avg_word < 50], num_bins, alpha = 0.5, facecolor = 'blue', edgecolor = 'black', label = 'disaster tweets') # Remove outliers
plt.hist(negative_avg_word[negative_avg_word < 50], num_bins, alpha = 0.5, facecolor = 'green', edgecolor = 'black', label = 'regular tweets') # Remove outliers
plt.xlabel("Average word length")
plt.ylabel("Tweet no")
plt.title("Average word length per tweet - train set")
plt.legend()
plt.show()
```

## Create the model

Extract the token count

```python
count_vectoriser = feature_extraction.text.CountVectorizer(ngram_range=(2,2)) #
tokens = count_vectoriser.fit_transform(train["text"])
```

```python
test.isnull().sum()
```

Fit the model

```python
model = linear_model.RidgeClassifier()
scores = model_selection.cross_val_score(model, tokens, train["target"], cv = 3, scoring = "f1")
model.fit(tokens, train["target"])
```
